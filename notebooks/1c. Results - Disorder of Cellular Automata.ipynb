Show results for specific Wolfram rules.

Uses a standard set of conditions of 100 cell width, 100 timestep depth, for 20 ensembles. Starting with an initial sparse layout, where only 10% of the cells start as 1 the rest are 0.

Optionally, save the image to the figures folder for including in the paper.

```python
def show_results(rule, save=False):
    ca = CA1DEnsemble(rule=rule, cells=100, count=20, init='sparse')
    ca.run(100) # timesteps
    ca.analyse()
    fig = ca.plot()
    
    labels = ['Divergence & Ergodic complexity for all time',
    'Divergence & Ergodic complexity for second half of steps (once stabalised)',
    'Divergence & Ergodic complexity for final step',
    'Divergence & Ergodic complexity of the diagonals Left to Right',
    'Divergence & Ergodic complexity of the diagonals Right to Left']
    divergence = ca.get_analysis('ergodic')-ca.get_analysis('ensemble').mean()
    complexity = ca.get_analysis('complexity')
    for i, k in enumerate(divergence):
        if i in [1,3,4]:
            print(divergence[k].mean(), complexity[k].mean(), labels[i])
    
    if save:
        fig.savefig("../figures/ca_%s.eps" % rule)
    return ca
```

## Ordered or mechanistic examples


```python
ca = show_results(rule=0)
```
