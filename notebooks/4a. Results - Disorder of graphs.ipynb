```python
import nbsetup
import numpy as np
import ergodicpy as ep
from simulations.graphs.models import ERGraph
```

```python
def series(nodes, ensembles=10):
    x, y = np.linspace(0, 1, 100), []
    # generate data
    for p in x:
        y.append([ERGraph(nodes, p).edges_counts() for _ in range(ensembles)])
    # analyse
    ees = ep.ErgodicSeries(x=x, y=y, x_label='p', bins=ep.binr(0,nodes,nodes))
    f = ees.results()
    return ees
```

```python
ees = series(10, 50)
```
