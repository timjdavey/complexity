```python
import nbsetup
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import pearsonr, spearmanr, kendalltau
import matplotlib.pyplot as plt
```

```python
try:
    funky = pd.read_csv("https://raw.githubusercontent.com/perlatex/R_for_Data_Science/master/demo_data/datasaurus.csv")
except:
    funky = pd.read_csv("../datasets/datasaurus.csv")
```

```python
plots = funky['dataset'].unique()[:6]
ratio = 3
fig, axes = plt.subplots(1, len(plots), sharex=False, sharey=False, figsize=(6*ratio,ratio))
for i, name in enumerate(plots):
    g = sns.scatterplot(data=funky.loc[funky.dataset == name], x='x', y='y', ax=axes[i], size=0.01, legend=False, color='black')
    g.set(xlabel=None, ylabel=None, xticklabels='', yticklabels='')
sns.despine()
fig.savefig('figures/dozen.eps')
```

```python
for name in plots:
    df = funky.loc[funky.dataset == name]
    print(name)
    print('mean', df.mean())
    print('var', df.var())
    print('pearson', pearsonr(df['x'], df['y']))
```
