```python
import nbsetup
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from simulations.daisy_world.model import POP_DEFAULT, DaisyWorld
```

```python
world = DaisyWorld(POP_DEFAULT, size=200)
world.plot()
world.simulate(20)
world.plot()
```

```python
def animate():
    world.step()
    return world.plot_grid()

fig = plt.figure()
anim = FuncAnimation(fig, animate, init_func=animate,
                               frames=20, interval=20, blit=True)
anim.save('daisyworld.gif')
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
plt.style.use('seaborn-pastel')


fig = plt.figure()
ax = plt.axes(xlim=(0, 4), ylim=(-2, 2))
line, = ax.plot([], [], lw=3)

def init():
    line.set_data([], [])
    return line,

def animate(i):
    x = np.linspace(0, 4, 1000)
    y = np.sin(2 * np.pi * (x - 0.01 * i))
    line.set_data(x, y)
    return line,

anim = FuncAnimation(fig, animate, init_func=init,
                               frames=200, interval=20, blit=True)


anim.save('sine_wave_2.gif')
from IPython.display import Image
Image(filename='sine_wave_2.gif')
```
