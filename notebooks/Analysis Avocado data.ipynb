```python
# developing reasonable bin strategy

print(df['AveragePrice'].min(), df['AveragePrice'].max())

# candidate price
print(df['Total Volume'].min(), df['Total Volume'].max())
```

```python
bins = np.linspace(df['AveragePrice'].min(),df['AveragePrice'].max(),20)
```

```python
observations = []
for r in df['region'].unique():
    vals = np.concatenate(df.loc[df['region'] == r].loc[:,['AveragePrice']].to_numpy())
    observations.append(vals)
observations = np.stack(observations)
```

```python
ee = ErgodicEnsemble(observations, bins)
```
